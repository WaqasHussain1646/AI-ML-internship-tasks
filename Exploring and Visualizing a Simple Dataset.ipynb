```python
# import labraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the dataset
df = pd.read_csv("iris.csv")

print("Dataset loaded successfully!")
```

```python
# 1. Print the shape (rows, columns)
print("Shape of the dataset:")
print(df.shape)
print("-" * 30)

# 2. Print column names
print("Column Names:")
print(df.columns)
print("-" * 30)

# 3. Print the first 5 rows
print("First 5 Rows:")
print(df.head())
print("-" * 30)
```

```python
# 1. Summary information (data types and non-null counts)
print("Dataset Info:")
print(df.info())
print("-" * 30)

# 2. Statistical summary (mean, std, min, max, etc.)
print("Statistical Description:")
print(df.describe())
```

```python
# draw scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', s=100)

plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
#  draw histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='petal_length', kde=True, bins=20, color='skyblue')

plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
#  draw box plot
plt.figure(figsize=(10, 6))
# Melts the dataset to long-form for easy plotting of all features side-by-side
sns.boxplot(data=df, orient="h", palette="Set2")

plt.title('Box Plot of Iris Features (Outlier Detection)')
plt.xlabel('Measurement (cm)')
plt.show()
```
